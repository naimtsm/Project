```python
import pandas as pd #pandas
import seaborn as sns # Data Visualisation
import matplotlib.pyplot as plt # Data Visualisation



#Retrieving HTML table data
url = 'https://www.basketball-reference.com/leagues/NBA_2019_per_game.html'
html = pd.read_html(url, header = 0)
df2019 =html[0]
df2019
```

```python
#Data Cleaning
raw = df2019.drop( df2019[df2019.Age == "Age"].index)
raw
#dop any header that redundant
```

#### DATA CLEANING

```python
raw.shape
```

```python
raw.head()
```

#### Check for any missing values

```python
raw.isnull().sum()
```

#### Replace missing values with 0

```python
df = raw.fillna(0)
```

```python
#check one more time missing value
df.isnull().sum()
#supposedly all zero because we already replace missing value with 0
```

```python
df
```

```python
#Dropping column Rk
df = df.drop(['Rk'], axis=1)
df
```

### Using Groupby() function

#### Which player scored the highest (PTS) in the Los Angeles Lakers?


```python
#Assigned new variable, LAL

LAL = df.groupby('Tm').get_group('LAL') #printing the all player of team LAL 
LAL
```

```python
#Highest scored (PTS)  player in Team LAL with PTS = 27.4
LAL[LAL.PTS == LAL.PTS.max()]
```

#### From the 5 positions, which position scores the most points?


######  Firstly, group the player by their position

```python
df.groupby('Pos').PTS.describe() 
#groupby position, the take the column points, PTS and use describe() function
```

##### Show only 5 positions that we wanted  

```python
positions = ['C', 'PF', 'SF', 'PG','SG'] #make a variable name positions, to list down 5 positions that we wanted. 

POS = df [df.Pos.isin(positions)] 
#isin being used to remove unwated position, only printed 5 positions that we declared at variable positions

#also can use this : POS =df [df['Pos'].isin(positions)]
POS
```

```python
POS.groupby('Pos').PTS.describe()
#From this table, we can conclude that the Position Centre, C have the highest mean which show that this position score the most points in 
```

```python
POINTS = df[['Pos','PTS']]
POINTS = POINTS[POINTS['Pos'].isin(positions)]

POINTS
```

### Scatter Plot

```python
sns.relplot(x="Pos", y="PTS", data=POINTS, hue = 'Pos')
#PG scored the highest value of points 
```

```python
sns.catplot(x="Pos", y="PTS", data=POINTS, jitter =0.4)
#
```

```python
myFigure= sns.catplot(x="Pos", y="PTS", kind="bar",data=POINTS) 

myFigure.fig.suptitle("Average points scored by each positions", fontsize=13)
```
